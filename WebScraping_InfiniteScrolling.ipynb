# Scraping all DeepAI content

This code downloads all the contents of the different URLs present in deepAI.org. The website implements infinite scroll load and hence the start and end of the number of pages have been noted in advance.

The URLs are first extracted and then content from each URL is saved into a PDF file. This PDF(s) can then be merged or referred to as different files.

```
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import pdfkit
```

```
base_url = []
for num in range(2,383):
    base_url.append(str("https://api.deepai.org/category_page_row_listing/definitions/popular/")+str(num))
```

```
pages = []
for url in base_url:
    page = requests.get(url)
    soup = bs(page.content,'html.parser')
    pages.append([a['href'] for a in soup.find_all('a',href=True) if a.text])
    print("Completed url ",url)
```

```
urls = []
for i in range(len(pages)):
    urls.append([set(i) for i in pages])
```

```
unique_list = [set(item) for sublist in urls for item in sublist]
urls_list1 = [item for sublist in unique_list for item in sublist]
urls_list2 = set(urls_list1)
urls_list3 = list(urls_list2)
urls_list3
```

```
final_url_list = ["http://deepai.org"+item for item in urls_list3]
final_url_list
```

```
filenames = []
for u in final_url_list:
    x = u.split('terms/',1)
    filenames.append(x[1])
```

```
for i in range(len(final_url_list)):
    pdfkit.from_url(final_url_list[i],"D://Personal//DeepAI//"+str(filenames[i])+".pdf",configuration=config)
```
